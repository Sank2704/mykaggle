```python
play.Sport.value_counts().plot()
#By default, plot() will plot the data as a line plot
```

```python
play.Sport.value_counts().plot(kind='bar')
#Kind parameter will give various options
#bar -- will plot a bar chart
#barh will plot the bar chart with horizontal axis
#pie will plot a pie chart
```

```python
play.Sport.value_counts().sort_values()[:10].plot(kind='pie')
#[:10] will limit the number of Sports displayed so that we can see clutter free plots
#the limiting numbers will work in all kind of plots
#use the ascending parameter inside the sort_values() to get the top or bottom 'n' numbers
```

```python
play.Sport.value_counts().sort_values()[:10].plot(kind='bar',color='red')
#Use the color attribute to use the color of your choice
# list of all the olors is availabe in ' https://matplotlib.org/examples/color/named_colors.html ' 
```

```python
#Function to label in plots
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.02*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
fig, ax = plt.subplots()
N = 10
total = play.Sport.value_counts().sort_values()[:N]
ind = np.arange(N)  # the x locations for the groups
width = 0.35  
ax.bar(ind + width, total, width, color='y')
for p in ax.patches:        
        ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.show()

#The above code how to get count of numbers in a bar plot
```
