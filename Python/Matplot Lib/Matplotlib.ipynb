```python
#Importing the Basic Package of Matplotlib
import matplotlib.pyplot as plt
#Importing Numpy package to do some basic operations
import numpy as np
import pandas as pd
%matplotlib inline
```

# Plot 1 (Basic Plot)

```python
x = np.linspace(0,10,100)
plt.plot(x,x,label='linear')
plt.legend()
plt.show()
```

```python
##Sub Plots
##Dividng the plot area , so that we can show multiple plots simultaneously 
##add_subplot(221), does the trick in ividing the total area
##(221) indicates the area to be a 2*2 division, and 1 represents the plot number
```

```python
fig = plt.figure()
ax = fig.add_subplot(221)
ax1 = fig.add_subplot(222)
ax.scatter(np.linspace(1,10,10), np.linspace(1,10,10))
ax1.scatter(np.linspace(1,20,20), np.linspace(1,20,20))
plt.show()
```

```python
##Labelling the Plots
```

```python
 x = [1,2,3,4]
y = [10,20,25,30]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, color='lightblue', linewidth=3)
ax.scatter(x,y,color='green',marker='^')
ax.set_xlim(1, 6.5)
ax.set(title = 'Sample Plot', xlabel='Numbers', ylabel='Range')
plt.show()
```

```python
#Saving the plots
```

```python
 x = [1,2,3,4]
y = [10,20,25,30]
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, color='lightblue', linewidth=3)
ax.scatter(x,y,color='green',marker='^')
ax.set_xlim(1, 6.5)
ax.set(title = 'Sample Plot', xlabel='Numbers', ylabel='Range')
plt.savefig('SamplePlot.png')
plt.savefig('SamplePlotTransparent.png',transparent=True)
plt.show()
```

```python
##Close and Cearing the plots
plt.cla()
plt.clf()
plt.close()
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
#Using ; at the end of plots avoids the description at the begining of the plot 
```

```python
#Two options for subplots
#1 Method
fig ,((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(np.random.random(10),np.random.random(10));
ax4.hist(np.random.random(10));
```

```python
#2 Method
fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].bar(np.random.random(10),np.random.random(10))
ax[1,1].hist(np.random.random(10));
```

```python
df1 = pd.read_csv('heart-disease (1).csv')
```

```python
#Disturbution showing the age
df1["age"].plot.hist(bins=50);
```

```python
df1.head()
```

```python
df1.plot.hist(figsize=(10,30),subplots=True);
#Not an ideal visualization
```

```python
above_50 = df1[df1["age"]>50]
above_50
```

```python
#PYPLOT Method
above_50.plot(kind="scatter",x="age",y='chol',c='target');
```

```python
##Object Oriented Method along with pyplot metod
fig,ax = plt.subplots(figsize=(10,6))
above_50.plot(kind='scatter',x="age",
              y="chol",c="target",
              ax=ax);
ax.set_xlim([45,100]) #Will set the initial and final limit for x-axis
```

```python
##Object oriented method

fig,ax= plt.subplots(figsize=(10,5))

scatter = ax.scatter(x=above_50["age"],y=above_50["chol"],
                        c=above_50["target"])

ax.set(title="Heart disease vs Cholestrol",xlabel="Age",
      ylabel="Cholestrol")

ax.legend(*scatter.legend_elements(),title="Target");

#Adding a horizotal line in the plot
ax.axhline(above_50["chol"].mean(),linestyle='--');
```

```python
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

scatter = ax0.scatter(x=above_50["age"],y=above_50["chol"],
                     c=above_50["target"])

ax0.set(title="Heart disease vs Cholestrol",
      ylabel="Cholestrol")

ax0.legend(*scatter.legend_elements(),title="Target");

ax0.axhline(above_50["chol"].mean(),linestyle='--');

#plotting to ax1
scatter = ax1.scatter(x=above_50["age"],y=above_50["thalach"],
                     c=above_50["target"]);

ax1.set(title="Heart disease vs thalach",xlabel="Age",
      ylabel="Thalah")

ax1.legend(*scatter.legend_elements(),title="Target");

ax1.axhline(above_50["thalach"].mean(),linestyle='--');
```
