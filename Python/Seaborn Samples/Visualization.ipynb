```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
hrdata = pd.read_csv('train_seaborn.xls')
votes = pd.read_csv('train1.xls')
```

```python
hrdata.head()
```

```python
votes.head()
```

```python
#Scatter plot 
sns.relplot(x="Views", y="Upvotes", data = votes)
```

```python
#Scatter plot with Hue
sns.relplot(x="Views", y="Upvotes", hue='Tag',data = votes)
```

```python
#Scatter plot with Hue
sns.relplot(x="Views", y="Upvotes", hue='Answers',data = votes)
```

```python
#Box plot
sns.catplot(x="education", y="avg_training_score", kind = "box", data=hrdata)
```

```python
#Box plot with specific Hue
sns.catplot(x="education", y="avg_training_score", hue = "is_promoted", kind = "box", data=hrdata)
```

```python
#Bar plot which shows the count of the values (with or without hue)
sns.catplot(x="education", y="avg_training_score", hue = "is_promoted", kind = "bar", data=hrdata)
```

# Visualizing the Distribution of a Dataset

```python
#Dist plot shows the distribution of a data wrt a numerical value
sns.distplot(hrdata.age)
```

```python
#Histogram, shows the distribution of data in the form of bins, where the number of bins control the bars count
sns.distplot(hrdata.age, kde=False, rug = True,bins=10)
```

There are various ways by which the plot area culd be customized
lets look at the most used ways below

```python
plt.figure(figsize=(10,10)) ## lets us specify the figure size
sns.barplot(x='age',y='length_of_service', data = hrdata)  #Our kind of plot
plt.xticks(rotation= 45) ## Lables in the x axis appear tilted, useful when we have large names 
plt.xlabel('Age') ## Lable for our x-axis
plt.ylabel('Experience') ##Lable for our Y-axis
plt.title('Age vs Service Exp comparison') #title for our axis
plt.show()  #Avoids the default plot text showin with the plot
```

```python
cor = hrdata.corr()
sns.heatmap(cor)
```
